```python
import numpy as np
import matplotlib.pyplot as plt

n = 52
F = np.zeros((n,n), dtype = complex)

w = np.e**(-2*np.pi*1j/n)

for i in range(n):
    for k in range(n):
        m = i*k
        F[i, k] = w**m

plt.imshow(F.real, cmap="jet")
plt.show
# plt.imshow(abs(F))
# plt.colorbar()
# plt.show()

x = np.random.randn(n)

X1 = F@x
X2 = np.fft.fft(x)
print(f'X1 == X2: {np.allclose(X1, X2)}')

# plot these
plt.figure()
plt.plot(abs(X1))
plt.plot(abs(X2), marker = 'o', color = 'r', linestyle = 'None')
plt.show()
plt.plot(F.real[:,25])
plt.show()
```
