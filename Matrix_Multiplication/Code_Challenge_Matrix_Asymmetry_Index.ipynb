# Code Challenge: Matrix Asymmetry Index

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Implement the code for computing Matrix Asymmetry Index

def MAI(A):
    Aanti = (A-A.T)/2
    mai = np.linalg.norm(Aanti) / np.linalg.norm(A)
    return mai
```

```python
m = 4

# Create a square matrix
A = np.random.randn(m, m)

# Create a (1) Symmetric (2) Skew-Symmetric and (3) Random Matrix from A
ASymm = (A+A.T)/2
ASkew = (A-A.T)/2
# A (for Random Matrix)

# Apply the Matrix Asymmetry Index using the norm of the created matrices
ai1 = MAI(ASymm)
ai2 = MAI(ASkew)
ai3 = MAI(A)

print(ai1)
print(" ")
print(ai2)
print(" ")
print(ai3)
```

```python

# Proportionally mixing the symmetric and skew-symmetric parts to form a matrix
# B = (1-p)*(A+A.T) + p*(A-A.T)

p = 1   # between 0 and 1
B = (1-p)*(A+A.T) + p*(A-A.T)
print(B)
print(" ")
ainew = MAI(B)
print(ainew)
```

```python
ps = np.linspace(0, 1, 100)
mai = np.zeros(len(ps))

for i in range(len(ps)):
    p = ps[i]
    A = np.random.randn(m, m)
    B = (1-p)*(A+A.T) + p*(A-A.T)
    mai[i] = MAI(B)

plt.plot(ps, mai, "-s")
plt.show()    
```
