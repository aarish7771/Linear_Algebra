# Code Challenge: Determinant of Matrix Product

```python
import numpy as np
import matplotlib.pyplot as plt 

# illustrate that det(AB) = det(A)*det(B)

# 1) for random 3x3 matrix

A = np.random.randn(3, 3)
B = np.random.randn(3, 3)
det = np.linalg.det(A@B)
detA = np.linalg.det(A)
detB = np.linalg.det(B)
print(det)
print(detA)
print(detB)
print(detA*detB)
print(det - detA*detB)

# 2) in a loop over random matrix up to 40x40

Dt = np.zeros((40, 2))

for i in range(40):
    A = np.random.randn(i, i)
    B = np.random.randn(i, i)
    d = np.linalg.det(A@B)
    dA = np.linalg.det(A)
    dB = np.linalg.det(B)
    Dt[i,0] = d
    Dt[i,1] = dA*dB

plt.plot(Dt[:,1] - Dt[:,0], 's-')
plt.ylim([-1,1])
plt.show()    
```
