# Code Challenge: Large Matrices with row exchanges

```python
import numpy as np

# generate a 6x6 matrix
 
A = np.random.randn(6, 6)
print(A)
print(" ")

# compute the determinant

print(np.linalg.det(A))
print(" ")

# one row exchange

A[:, [0, 1]] = A[:, [1, 0]]
print(A)
print(" ")
print(np.linalg.det(A))
print(" ")

# two row exchange

A[:, [1, 2]] = A[:, [2, 1]]
print(A)
print(" ")
print(np.linalg.det(A))
```
