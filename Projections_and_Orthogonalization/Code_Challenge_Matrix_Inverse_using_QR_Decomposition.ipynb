# Code Challenge: Matrix Inverse using QR Decomposition

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a large (e.g- N=100) matrix and compute it's inverse using QR Decomposition and inv()

x = 100
A = np.random.randn(x, x)

# Explicit Inverse
Ainv = np.linalg.inv(A)

# Inverse via QR
(Q,R) = np.linalg.qr(A)
plt.imshow(R)
plt.show()
```

```python
Ainvqr1 = np.linalg.inv(R)@Q.T

plt.subplot(1,2,1)
plt.imshow(Ainvqr1)
plt.title('A$^{-1}$ from inv')

Ainvqr2 = np.linalg.solve(R, Q.T)

plt.subplot(1,2,2)
plt.imshow(Ainvqr2)
plt.title('A$^{-1}$ from solve')

plt.show()
```

```python
np.corrcoef((Ainv.flatten(), Ainvqr1.flatten(), Ainvqr2.flatten()))
```
