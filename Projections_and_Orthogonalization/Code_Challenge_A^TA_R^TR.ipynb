# Code Challenge: A^TA = R^TR (R from QR Decomposition)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1) Generate random matrix A
# 2) Compute its QR Decomposition
# 3) Test the claim
# 4) Prove it using a paper-and-pen

m = 10
n = 4
A = np.random.randn(m,n)

(Q,R) = np.linalg.qr(A)

print(A.T@A)
print(" ")
print(R.T@R)
print(" ")

plt.subplot(1,2,1)
plt.imshow(A.T@A)

plt.subplot(1,2,2)
plt.imshow(R.T@R)

plt.show()
```
