# Code Challenge: Shermann Morrison Formula 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 5
a = np.random.randn(m, 1)
b = np.random.randn(m, 1)

I = np.eye(5)
A = I - a@b.T
Ainv = I + (a@b.T)/(1 - np.dot(a.T, b))

# Check whether A@Ainv is an Identity Matrix
print(A@Ainv)
plt.imshow(A@Ainv)
plt.show()
```
