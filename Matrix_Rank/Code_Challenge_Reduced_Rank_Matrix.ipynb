# Code Challenge: Reduced-Rank Matrix

# Create Reduced Rank Matrix using Matrix Multiplication

# Create a 10x10 matrix with rank = 4 (via matrix multiplication)

# Generalize this procedure to create any MxN matrix with rank = R

```python
import numpy as np

# Create a 10x10 matrix with rank = 4 (via matrix multiplication)

A = np.random.randn(10, 4)
B = np.random.randn(4, 10)
C = A@B
print(C)
r = np.linalg.matrix_rank(C)
print(r)
```

```python
# Generalize this procedure to create any MxN matrix with rank = R

M = 9
N = 7
R = 5

A = np.random.randn(M, R)
B = np.random.randn(R, N)
C = A@B
print(C)
r = np.linalg.matrix_rank(C)
print(r)
```
